```python
from pathlib import Path
import json
from datetime import datetime
import os
from pyspark.sql import SparkSession, functions as F
```

```python
BASE = Path("/home/jovyan/work")
INBOX = BASE / "data" / "inbox"
REFERENCE = BASE / "data" / "reference"
OUTBOX = BASE / "data" / "outbox" / "trips_enriched.parquet"
STATE = BASE / "state"
MANIFEST = STATE / "manifest.json"
LOOKUP = REFERENCE / "taxi_zone_lookup.parquet"
SPARK_EVENTS = STATE / "spark-events"

print("INBOX:", INBOX)
print("OUTBOX:", OUTBOX)
print("MANIFEST:", MANIFEST)
print("LOOKUP:", LOOKUP)
```

```python
os.makedirs(SPARK_EVENTS, exist_ok=True)

spark = (
    SparkSession.builder
    .appName("proj1")
    .master("local[*]")
    .config("spark.sql.shuffle.partitions", "8")
    .config("spark.eventLog.enabled", "true")
    .config("spark.eventLog.dir", f"file:{SPARK_EVENTS}")
    .getOrCreate()
)
```

```python
if MANIFEST.exists():
    with open(MANIFEST, "r") as f:
        manifest = json.load(f)
else:
    manifest = {"processed_files": []}

processed_files = {entry["filename"] for entry in manifest["processed_files"]}

trip_files = sorted(INBOX.glob("*.parquet"))
new_files = [f for f in trip_files if f.name not in processed_files]

print("Processed:", processed_files)
print("New:", [f.name for f in new_files])
```

```python
def transform(trips_df):    
    raw_row_count = trips_df.count()
    # parse and cast types correctly
    # parquet retains types, but this enforces standard sizing and catches schema drift
    trips_transformed = (
        trips_df
        .withColumn("VendorID", F.col("VendorID").cast("integer"))
        .withColumn("PULocationID", F.col("PULocationID").cast("integer"))
        .withColumn("DOLocationID", F.col("DOLocationID").cast("integer"))
        .withColumn("passenger_count", F.col("passenger_count").cast("integer"))
        .withColumn("trip_distance", F.col("trip_distance").cast("double"))
        .withColumn("fare_amount", F.col("fare_amount").cast("double"))
        .withColumn("total_amount", F.col("total_amount").cast("double"))
        .withColumn("tpep_pickup_datetime", F.to_timestamp(F.col("tpep_pickup_datetime")))
        .withColumn("tpep_dropoff_datetime", F.to_timestamp(F.col("tpep_dropoff_datetime")))
    )

    # apply data cleaning rules
    # rule 1: drop rows with missing critical IDs and timestamps
    # rule 2: ensure trip_distance > 0 (filters out system errors or canceled rides)
    # rule 3: ensure total_amount >= 0 (filters out negative charge refunds/glitches)
    trips_cleaned = (
        trips_transformed
        .dropna(subset=["VendorID", "PULocationID", "DOLocationID", "tpep_pickup_datetime", "tpep_dropoff_datetime"])
        .filter(F.col("trip_distance") > 0.0)
        .filter(F.col("total_amount") >= 0.0)
    )

    rows_after_cleaning = trips_cleaned.count()

    # deduplicate records using a defined key
    # TLC data lacks a primary unique key. use a composite key of the vendor, pickup/dropoff times, and pickup/dropoff locations. 
    # the likelihood of the exact same taxi making two separate trips at the exact same second between the exact same zones is effectively zero.
    dedup_key = ["VendorID", "tpep_pickup_datetime", "tpep_dropoff_datetime", "PULocationID", "DOLocationID"]
    trips_final = trips_cleaned.dropDuplicates(subset=dedup_key)
    
    rows_after_dedup = trips_final.count()
    duplicates_dropped = rows_after_cleaning - rows_after_dedup

    metrics = {
        "raw_row_count": raw_row_count,
        "rows_after_cleaning": rows_after_cleaning,
        "rows_after_dedup": rows_after_dedup,
        "duplicates_dropped": duplicates_dropped
    }

    return trips_final, metrics
```

# Enrichment

```python
zones_df = (
    spark.read.parquet(str(LOOKUP))
    .select(
        F.col("LocationID"),
        F.col("Zone"),
        F.col("Borough"),
        F.col("service_zone")
    )
)
```

```python
pu_zones = zones_df.select(
    F.col("LocationID").alias("pu_location_id_lookup"),
    F.col("Zone").alias("pickup_zone_name"),
    F.col("Borough").alias("pickup_borough"),
    F.col("service_zone").alias("pickup_service_zone")
)

do_zones = zones_df.select(
    F.col("LocationID").alias("do_location_id_lookup"),
    F.col("Zone").alias("dropoff_zone_name"),
    F.col("Borough").alias("dropoff_borough"),
    F.col("service_zone").alias("dropoff_service_zone")
)
```

```python
run_manifest_entries = []
new_enriched_dfs = []

for f in new_files:
    raw_df = (
        spark.read.parquet(str(f))
        .withColumn("source_file", F.input_file_name())
        .withColumn("ingested_at", F.current_timestamp())
    )

    trips_final, metrics = transform(raw_df)

    enriched_df = (
        trips_final
        .join(
            F.broadcast(pu_zones),
            trips_final["PULocationID"] == pu_zones["pu_location_id_lookup"],
            "left"
        )
        .join(
            F.broadcast(do_zones),
            trips_final["DOLocationID"] == do_zones["do_location_id_lookup"],
            "left"
        )
        .drop("pu_location_id_lookup", "do_location_id_lookup")
    )

    before_count = trips_final.count()
    after_count = enriched_df.count()

    print(f"File: {f.name}")
    print("Before enrichment:", before_count)
    print("After enrichment:", after_count)

    enriched_df.select(
        "PULocationID",
        "pickup_zone_name",
        "DOLocationID",
        "dropoff_zone_name"
    ).show(10, truncate=False)

    enriched_df.select(
        F.sum(F.col("pickup_zone_name").isNull().cast("int")).alias("missing_pickup_zones"),
        F.sum(F.col("dropoff_zone_name").isNull().cast("int")).alias("missing_dropoff_zones")
    ).show()

    new_enriched_dfs.append(enriched_df)

    run_manifest_entries.append({
        "filename": f.name,
        "file_size": f.stat().st_size,
        "processed_at": datetime.utcnow().isoformat(),
        **metrics
    })

if run_manifest_entries:
    summary_df = spark.createDataFrame(run_manifest_entries)
    print("Run summary:")
    summary_df.show(truncate=False)
else:
    print("No new files processed on this run.")
```

```python
# output
if new_enriched_dfs:
    # 1. combine all newly processed DataFrames
    final_new_df = new_enriched_dfs[0]
    for df in new_enriched_dfs[1:]:
        final_new_df = final_new_df.unionByName(df)

    # 2. add derived columns
    # use unix_timestamp to get the difference in seconds, then divide by 60
    final_new_df = final_new_df.withColumn(
        "trip_duration_minutes",
        (F.unix_timestamp("tpep_dropoff_datetime") - F.unix_timestamp("tpep_pickup_datetime")) / 60.0
    ).withColumn(
        "pickup_date",
        F.to_date("tpep_pickup_datetime")
    )

    # 3. select only the required fields (and any others you want to keep)
    # this ensures the schema remains consistent for the output dataset
    output_df = final_new_df.select(
        "tpep_pickup_datetime",
        "tpep_dropoff_datetime",
        "PULocationID",
        "DOLocationID",
        "pickup_zone_name",
        "dropoff_zone_name",
        "passenger_count",
        "trip_distance",
        "trip_duration_minutes",
        "pickup_date",
        "source_file",
        "ingested_at"
    )

    # 4. write to the outbox
    # 'append' mode allows the dataset to grow with each run.
    # idempotency is guaranteed by the manifest check at the start of the script.
    print(f"Writing {output_df.count()} new records to {OUTBOX}...")
    
    output_df.write.mode("append").parquet(str(OUTBOX))
    
    print("Write complete.")
else:
    print("No new data to write.")
```

# Manifest update

```python


# Manifest update
# Run this only after the final output dataset has been written successfully.

manifest_entries = manifest["processed_files"] + run_manifest_entries

with open(MANIFEST, "w") as fp:
    json.dump({"processed_files": manifest_entries}, fp, indent=2)

print(f"Manifest updated with {len(run_manifest_entries)} new files.")
```
