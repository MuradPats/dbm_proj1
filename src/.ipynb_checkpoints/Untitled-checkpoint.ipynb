```python
from pathlib import Path
import json
from datetime import datetime

BASE = Path("/home/jovyan/work")
INBOX = BASE / "data" / "inbox"
REFERENCE = BASE / "data" / "reference"
OUTBOX = BASE / "data" / "outbox" / "trips_enriched.parquet"
STATE = BASE / "state"
MANIFEST = STATE / "manifest.json"
LOOKUP = REFERENCE / "taxi_zone_lookup.parquet"
SPARK_EVENTS = STATE / "spark-events"

print("INBOX:", INBOX)
print("OUTBOX:", OUTBOX)
print("MANIFEST:", MANIFEST)
print("LOOKUP:", LOOKUP)
```

```python
import os
from pathlib import Path
from pyspark.sql import SparkSession, functions as F


os.makedirs(SPARK_EVENTS, exist_ok=True)

spark = (
    SparkSession.builder
    .appName("proj1")
    .master("local[*]")
    .config("spark.sql.shuffle.partitions", "8")
    .config("spark.eventLog.enabled", "true")
    .config("spark.eventLog.dir", f"file:{SPARK_EVENTS}")
    .getOrCreate()
)
```

```python
trips = spark.read.parquet(str(INBOX / "yellow_tripdata_2025-01.parquet"))
zones = spark.read.parquet(str(LOOKUP))

trips.printSchema()
zones.show(5, truncate=False)
```

```python
import json

if MANIFEST.exists():
    with open(MANIFEST, "r") as f:
        manifest = json.load(f)
else:
    manifest = {"processed_files": []}

processed_files = {entry["filename"] for entry in manifest["processed_files"]}

trip_files = sorted(INBOX.glob("*.parquet"))
new_files = [f for f in trip_files if f.name not in processed_files]

print("Processed:", processed_files)
print("New:", [f.name for f in new_files])
```

```python
if new_files:
    trips_df = (
        spark.read.parquet(*[str(f) for f in new_files])
        .withColumn("source_file", F.input_file_name())
        .withColumn("ingested_at", F.current_timestamp())
    )
    
    print("New input rows:", trips_df.count())
    trips_df.printSchema()
else:
    print("No new files to process.")
```
